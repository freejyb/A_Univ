```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#데이터 생성
df1=pd.DataFrame({'X':[10,24,25,40,55],'Y':[150,160,170,180,177]})

#그래프생성

# figure 는 그래프가 그려지는 화면--> subplot 은 그래프이고,  한 figure 안에 여러개의 그래프를 그릴 수 있다. 
# 그래프는 보통 ax 라고 명명

fig,ax=plt.subplots(2,2)

ax[0,0].plot(df1['X'],df1['Y'],color='blue',linestyle='solid',marker='o')
ax[0,1].plot(df1['X'],df1['Y'],color='blue',linestyle='dotted',marker='o')
ax[1,0].plot(df1['X'],df1['Y'],color='blue',linestyle='dashed',marker='o')
ax[1,1].plot(df1['X'],df1['Y'],color='blue',linestyle='dashdot',marker='o')

#그래프 제목 추가
ax[0, 0].set_title("solid")
ax[0, 1].set_title("dotted")
ax[1, 0].set_title("dashed")
ax[1, 1].set_title("dashdot")

#간격설정
plt.subplots_adjust(wspace=0.2, hspace=0.4)

#그래프 출력
plt.show()
```

## 범례 표시하기( legend)

```python
# Legend, 차트의 범례 그리기 ( 여러개의 챠트가 존재하는 경우 설명표시를 하고픈 경우) 
# 그래프마다 라벨을 달고 이 라벨명을 출력할 수 있는 것
# legend()를 이용하여 각 데이터가 무엇인지 설명해주는 Legend(범례) 가능

# legend() : 차트에 각 데이터의 설명 추가

# 구글 : matplotlib 범례 혹은 matplotlib legend를 검색
#  plt. 뒤에 붙어있는 기능 중 범례로 언급되는 기능 이름을 찾고,
# 예제로 이동해서 plt.legend()를 찾는다.

# legend를 추가하려면 .plot 안에 label = "표시될 내용"을 추가해야한다.

## 레전드 : 그래프를 그릴때 여러개의 그래프를 같이 그리는 경우
# legend를 사용하기 위해서는 plt.plot에서 label 변수에 그래프의 이름을 정의
# plt.legend(‘위치')를 정함

# 범례 위치 결정해주기
# best
# upper right
# upper left
# lower left
# lower right
# right
# center left
# center right
# lower center
# upper center
# center
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10,0.1)
y = x*0.2
y2 = np.sin(x)

plt.plot(x,y,'b',label='first') # 그래프 이름 정의
plt.plot(x,y2,'r',label='second') # 그래프 이름 정의
plt.xlabel('x axis') 
plt.ylabel('y axis')
plt.title('matplotlib sample')# 그래프 제목 정의
plt.legend(loc='upper left')#  plt.legend(‘위치') 정의
plt.show()
```

# 서브플롯(subplot)

```python
## 서브플롯 :   여러개의 그래프를 그리는 경우 사용
# 서브플롯은 그래프가 그려질 위치를 격자형으로 지정
# ==> plt.subplot(nrow,ncol,pos) 입력 사용 : 행의 개수, 열의 개수, 차트 위치 순으로 입력

# nrow,:  dataframe의 행의 수를 출력하는 함수
# ncol : column(열)의 개수 --> 변수의 개수를 세주는 함수
# pos : 현재 위치(좌표)
# nrow,ncol ==> 3,2 이면 3줄로, 가로는 2칸으로 된 그래프 plain 설정
# pos는 몇번째 plain에 그래프를 그릴지 지정하는데, 상단에서 부터 우측,아래 방향으로 1,2,3,4,5,6 순서
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)
y1 = x*5
y2 = x*10
y3 = x*3
y4 = x*1

plt.subplot(3,1,1)
plt.plot(x,y1)
plt.subplot(3,1,1)
plt.plot(x,y2)
plt.subplot(3,1,1)
plt.plot(x,y3)
plt.subplot(3,1,1)
plt.plot(x,y4)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)
y1 = x*5
y2 = x*10
y3 = x*3
y4 = x*1

plt.subplot(2,1,1)  # 1면에
plt.plot(x,y1)
plt.subplot(2,1,1)
plt.plot(x,y2)
plt.subplot(2,1,1)
plt.plot(x,y3)
plt.subplot(2,1,1)
plt.plot(x,y4)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)

y1 = x*5
y2 = x*10
y3 = x*3
y4 = x*1
plt.subplot(2,1,1)  #  2,1 크기의 plain 을 만들어놓고 그래프를 위,아래로 두개를 그리는 경우
plt.plot(x,y1)
plt.subplot(2,1,2) # 두번째 면에
plt.plot(x,y2)
plt.subplot(2,1,2) # 두번째 면에
plt.plot(x,y3)
plt.subplot(2,1,2) # 두번째 면에
plt.plot(x,y4)
plt.show()
```

```python
 # 2x2 plain으로 4개의 그래프를 그린 경우
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,20)
y1 = x*5
y2 = x*1
y3 = x*2
y4 = x*3
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```

```python
# 그래프 사이즈 변경
# plt.figure를 이용 figsize=(가로,세로)를 인자로 크기 조절

# figure 는 그래프가 그려지는 화면--> subplot 은 그래프이고,  한 figure 안에 여러개의 그래프를 그릴 수 있다. 
# 그래프는 보통 ax 라고 명명
```

```python
import numpy as np
x = np.arange(1,20)
y1 = x*5
y2 = x*1
y3 = x*2
y4 = x*3
plt.figure(figsize=(10,5)) # 10x5 크기로 그래프 크기를 지정한 경우
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```

```python
height = np.random.normal(size = 1000)


# 분할 구간의 개수를 지정해서 나누기
plt.subplot(3,1,1)
plt.hist(height, bins = 4)

# 구간을 자동으로 분할하기
plt.subplot(3,1,2)
plt.hist(height, bins = 'auto')

# 막대 넓이 조절
plt.subplot(3,1,3)
plt.hist(height, bins = 'auto', rwidth = 0.8)
```

# 막대 그래프 (Bar chart)

```python
## 막대 그래프 (Bar chart)
# 히스토그램과 비슷한 형식의 막대 그래프
# plt.bar()를 이용 
# plt.hist()는 연속적인 데이터를 표현하고 plt.bar()는 카테고리를 나눌 수 있는 데이터를 표현
# plt.barh()를 이용하면 가로 막대 형식으로 출력 가능
# plt.bar() : 세로 막대 차트
```

```python
x = ['서울', '부산', '대구', '인천경찰청'] # x values
y = [32, 123, 53, 11] # height

plt.bar(x,y)
plt.title('police')
plt.xlabel('지방경찰청')
plt.ylabel('검거실적')
plt.legend(['지방경찰정별 검거실적'])
plt.show()
```

```python
# seaborn  사용

sns.barplot(x=x,y=y)
plt.title('police')
plt.xlabel('지방경찰청')
plt.ylabel('검거실적')
plt.legend(['지방경찰정별 검거실적'])
plt.show()
```
