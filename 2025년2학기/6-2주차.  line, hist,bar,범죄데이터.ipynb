# 서브플롯(subplot)

```python
## 서브플롯 :   여러개의 그래프를 그리는 경우 사용
# 서브플롯은 그래프가 그려질 위치를 격자형으로 지정
# ==> plt.subplot(nrow,ncol,pos) 입력 사용 : 행의 개수, 열의 개수, 차트 위치 순으로 입력

# nrow,:  dataframe의 행의 수를 출력하는 함수
# ncol : column(열)의 개수 --> 변수의 개수를 세주는 함수
# pos : 현재 위치(좌표)
# nrow,ncol ==> 3,2 이면 3줄로, 가로는 2칸으로 된 그래프 plain 설정
# pos는 몇번째 plain에 그래프를 그릴지 지정하는데, 상단에서 부터 우측,아래 방향으로 1,2,3,4,5,6 순서
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)
y1 = x*5
y2 = x*10
y3 = x*3
y4 = x*1

plt.subplot(3,1,1)
plt.plot(x,y1)
plt.subplot(3,1,1)
plt.plot(x,y2)
plt.subplot(3,1,1)
plt.plot(x,y3)
plt.subplot(3,1,1)
plt.plot(x,y4)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)

y1 = x*5
y2 = x*10
y3 = x*3
y4 = x*1
plt.subplot(2,1,1)  #  2,1 크기의 plain 을 만들어놓고 그래프를 위,아래로 두개를 그리는 경우
plt.plot(x,y1)
plt.subplot(2,1,2) # 두번째 면에
plt.plot(x,y2)
plt.subplot(2,1,2) # 두번째 면에
plt.plot(x,y3)
plt.subplot(2,1,2) # 두번째 면에
plt.plot(x,y4)
plt.show()
```

```python
 # 2x2 plain으로 4개의 그래프를 그린 경우
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,20)
y1 = x*5
y2 = x*1
y3 = x*2
y4 = x*3
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```

```python
# 그래프 사이즈 변경
# plt.figure를 이용 figsize=(가로,세로)를 인자로 크기 조절

# figure 는 그래프가 그려지는 화면--> subplot 은 그래프이고,  한 figure 안에 여러개의 그래프를 그릴 수 있다. 
# 그래프는 보통 ax 라고 명명
```

```python
import numpy as np
x = np.arange(1,20)
y1 = x*5
y2 = x*1
y3 = x*2
y4 = x*3
plt.figure(figsize=(10,5)) # 10x5 크기로 그래프 크기를 지정한 경우
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```
