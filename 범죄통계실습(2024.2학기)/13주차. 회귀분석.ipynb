# 회귀분석(Regression Analysis)

## 회귀분석 이해

```python
# 하나 이상의 독립 변수(예측 변수)가 종속 변수(결과 변수)에 미치는 영향을 설명하고 예측하는 통계적 기법입니다. 
# 독립변수(설명변수, 피쳐)와 종속변수간의 인과관계 분석 ==> 회귀계수 ( 독립변수 --> 종속변수)
# 회귀분석을 통해 변수들 간의 관계를 수량화하고, 주어진 독립 변수를 사용해 종속 변수를 예측할 수 있습니다. 
# 회귀분석은 데이터 분석, 머신러닝, 경제학, 생물학 등 다양한 분야에서 활용됩니다.

# 회귀분석의 주요 목적

# 1. 변수 간의 관계를 설명: 독립 변수와 종속 변수 사이의 관계를 파악하여 데이터의 패턴을 이해합니다.
# 2. 예측: 과거 데이터로 학습한 모델을 사용해 새로운 데이터에서 종속 변수를 예측합니다.
# 3. 변수의 영향력 평가: 독립 변수가 종속 변수에 미치는 영향력을 평가하고 중요한 변수를 파악합니다.
```

## 선형회귀분석(Linear Regression) 구분

```python
# 선형 회귀분석(Linear Regression)은 가장 기본적이고 널리 사용되는 통계적 기법 중 하나
# 독립 변수(또는 예측 변수)와 종속 변수(또는 결과 변수) 사이의 선형적인 관계를 모델링하는 방법입니다
# 선형 회귀의 주요 목적 : 독립 변수를 사용해 종속 변수를 예측하는 것과 두 변수 간의 관계를 설명하는 것입니다.
```

###  단순 선형회귀 (Simple Linear Regression)

```python
# 단순 선형회귀 (Simple Linear Regression)
# 단순 선형회귀는 하나의 독립 변수와 하나의 종속 변수 사이의 선형 관계를 모델링하는 기법입니다. 
# 회귀식은 다음과 같은 형태를 가집니다:

# y = ax + b
# 단순회귀 : 독립변수(x)가 1개 
# 단순회귀 가정 : 선형성 linear (x변수와 y변수간)  f(x1 + x2) = f(x1) + f(x2)  , f(ax) = af(x)
```

### 다중 선형회귀 (Multiple Linear Regression)

```python
# 다중 선형회귀 (Multiple Linear Regression)
# 다중 선형회귀는 두 개 이상의 독립 변수가 종속 변수에 미치는 영향을 설명하는 모델입니다. 
```

## 사례 보기

```python
# 단순회귀분석(Simple Linear Regression)은 하나의 독립 변수와 하나의 종속 변수 간의 관계를 분석하는 데 사용됩니다. 
# statsmodels 라이브러리를 사용하여 단순회귀분석을 수행하는 사례를 보여드리겠습니다. 
```

```python
# 여기서는 인구수(Population, 독립변수)가 범죄건수(Crime, 종속변수)에 미치는 영향을 분석합니다
```

### 데이터 준비

```python
# 'Crime': [10, 20, 15, 30, 25, 40, 35, 50, 45, 60],
#  Population': [1000, 1500, 1200, 2000, 1700, 2500, 2300, 3000, 2800, 3500]
```

### 라이브러리

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

### 단순회귀분석

```python
# 가상의 데이터 생성
data = {
    'Crime': [10, 20, 15, 30, 25, 40, 35, 50, 45, 60],
    'Population': [1000, 1500, 1200, 2000, 1700, 2500, 2300, 3000, 2800, 3500]
}

df = pd.DataFrame(data)

# 독립변수와 종속변수 설정
X = df['Population']
y = df['Crime']

# 상수항 추가
X = sm.add_constant(X)

# 회귀모델 생성
model = sm.OLS(y, X).fit()

# 모델 결과 요약
summary = model.summary()
print(summary)
```

### 결과 해석

```python
# R-squared (결정계수): 0.999

# 이 값은 모델이 종속 변수의 변동성을 99.9% 설명하고 있음을 나타냅니다. 매우 높은 설명력을 갖고 있습니다.
# F-statistic: 5783

# 모델의 전체 유의성을 평가하는 F-통계량입니다. 매우 큰 값으로, 모델이 통계적으로 유의미함을 나타냅니다.
# Prob (F-statistic): 9.97e-13

# F-통계량에 대한 p-value입니다. 0.05보다 훨씬 작아서, 모델이 유의미하다고 결론지을 수 있습니다.

# 회귀 계수 (coefficients)

# const (절편): 0.0011
# 회귀 직선이 Y축과 만나는 지점입니다. 이 값은 거의 0에 가까운 작은 값입니다.
# Population (인구수): 0.0196
# 인구수가 1명 증가할 때 범죄건수 0.0196만큼 증가함을 의미합니다. 이는 통계적으로 유의미한 관계를 나타냅니다 (p-value < 0.05).
```

```python
# 결론
# 이 단순회귀분석 결과에서 인구수는 범죄건수아 통계적으로 유의미한 영향을 미치는 것으로 나타났습니다. 
# 인구수가 증가할수록 범죄건수도 증가하는 경향이 있음을 알 수 있습니다. 
```

## Statsmodel 라이브러리
https://www.statsmodels.org/stable/index.html

```python
# 분석에 앞서 먼저 시각화를 이용하여 상관성을 육안으로 확인해본다.
# 단순회귀분석을 위해 나이와 범죄건수에 대한 단순회귀분석 모델 구현
# 상관분석에 앞서 먼저 시각화를 이용하여 상관성을 육안으로 확인해본다.


# 상관분석에 앞서 먼저 시각화를 이용하여 상관성을 육안으로 확인해본다.
sub = data3[['전체형법범죄', '살인']]
# # 각 데이터의 히스토그램과 산점도를 도출(산점도에는 회귀선까지 구현)
sns.pairplot(sub, kind='reg')
plt.show()
```

```python
# 필요한 라이브러리
import statsmodels.api as sm

# target은 종속변수, feature는 독립변수와 같다.
target = data3['전체형법범죄']
feature = data3['살인']

# 상수항을 더해줘서 회귀식을 더 잘 적합할 수 있도록 만들어 준다.
feature_add = sm.add_constant(feature, has_constant='add')
# 상수항 더한 것 확인
# feature_add
```

```python
# sm OLS 적합
model = sm.OLS(target, feature_add)
fitted_model = model.fit()

# summary 함수통해 결과출력
fitted_model.summary()
```

```python
# 통계치 출력
print(f'결정계수 R-squared : {fitted_model.rsquared : .5f}')
print(f'constant coef(계수)에 대한 p 값 : {fitted_model.pvalues[0] : .5f}')
print(f'feature coef(계수)에 대한 p 값 : {fitted_model.pvalues[1] : .5f}')
print(f'잔차의 제곱합 MSE : {fitted_model.mse_resid : .5f}')

# 회귀선 도출
target_pred1 = fitted_model.predict()

plt.scatter(feature, target, marker='+', label='$Y_i$ 실제 관측치')
plt.plot(feature, target_pred1, color = 'g', label='$\hat{Y}_i$ 회귀선')
plt.legend()
plt.xlabel('살인')
plt.ylabel('전체형법범죄')
plt.show()
```

## Sklearn 라이브러리

```python
# sklearn을 활용한 선형회귀선 도출
from sklearn.linear_model import LinearRegression

model2 = LinearRegression()
model2.fit(feature_add, target)

target_pred2 = model2.predict(feature_add)
```

```python
print(f'결정계수 R-squared : {model2.score(feature_add, target) : .5f}')
print(f'constant coef(계수): {model2.intercept_ : .5f}')
print(f'feature coef(계수) : {model2.coef_[1] : .5f}')

plt.scatter(feature, target, marker='+', label='$Y_i$ 실제 관측치')
plt.plot(feature, target_pred2, color = 'g', label='$\hat{Y}_i$ 회귀선')
plt.legend()
plt.xlabel('살인')
plt.ylabel('전체형법범죄')
plt.show()
```

## Scipy 라이브러리

```python
from scipy.stats import linregress

model = linregress(feature, target)

print(f"R-squared : {model.rvalue ** 2 : .5f}")
print(f'constant coef(계수) : {model.intercept : .5f}')
print(f'feature coef(계수) : {model.slope : .5f}')
print(f'feature coef(계수)의 p-value : {model.pvalue : .5f}')
```

```python
plt.plot(feature, target, '+', label='$Y_i$ 실제 관측치')
plt.plot(feature, model.intercept + model.slope * feature, 'g', label='$\hat{Y}_i$ 회귀선')
plt.xlabel('살인')
plt.ylabel('전체형법범죄')
plt.legend()
plt.show()
```
