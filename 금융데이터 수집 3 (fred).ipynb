##### 미국 GDP(GDP)

```python
# 미국 GDP(GDP)
import pandas_datareader as web

data1 = web.DataReader('GDP', 'fred', '1990-01-01')
data1.head()
```

```python
import matplotlib.pyplot as plt
data1.plot(figsize=(10, 6))
plt.show()
```
