#### 1. 공분산 (Covariance)

```python
# 의미 :  두 변수 간 함께 변화하는 동일 방향성 여부를 파악하는 값

# 공분산이 양(+)인 경우: 두 변수가 동일 방향으로 변화(하나가 증가하면 다른 하나도 증가)
# 공분산이 음(-)인 경우: 두 변수가 반대 방향으로 변화(하나가 증가하면 다른 하나는 감소)
```

```python
# 넘파이 이용
# 계산 : np.cov(변수명)
```

```python
m = [1, 3, 5, 7]
n = [10, 16, 24, 30]

import numpy as np

np.cov(m, n)
```

```python
# 공분산만 보고자 하는 경우

np.cov(m, n)[0, 1]
```

#### 3. 상관계수 (correlation coefficient)

```python
# 두 변수 간 변화 정도
# -1 ~ 1 범위의 수로 나타낸 것
```

#### 1) 피어슨 상관계수

```python
# 칼 피어슨(Karl Pearson)이 개발한 상관계수
# 적률상관계수(product-moment correlation coefficient)
# 연속형 변수 간 상관분석
# 등간척도/비율척도 를 사용한 변수에 적용가능
```

```python
# np.corrcoef(변수명, 변수명) 함수로 계산
```

```python
import numpy as np

np.corrcoef(m, n)
```

```python
# 두 변수 간 상관계수만 확인

np.corrcoef(m, n)[0, 1]
```

#### 2) Spearman 상관계수 

```python
# 명목척도 를 사용한 변수 간 상관분석
# 등간/비율척도를 사용한 변수들이라 하더라도, 두 변수 간의 관계가 비선형적 일 때 상관분석에 활용
```

```python
import scipy.stats

scipy.stats.spearmanr(m, n).correlation
```

#### 3) kendall의 tau 

```python
# 비선형적 관게이거나 서열변수일 때 사용
```

```python
scipy.stats.kendalltau(m, n).correlation
```

#### 4. 상관계수 통계적 검증 

```python
# 상관계수는 표본에서 구한 것
# 표본 상관계수에 대한 통계적 유의성 검증이 필요
# 두 변수의 피어슨 상관계수와 p 값을 계산
# p 값 : 유의확률  p< 0.05(유의수준) 인 경우 귀무가설을 기각(상관계수= 0)하고 , 대립가설을 채택

# 반환값 : ( 상관계수 값 , p-value)

import scipy.stats

scipy.stats.pearsonr(m, n)
```
